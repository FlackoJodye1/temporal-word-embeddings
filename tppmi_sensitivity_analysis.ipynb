# TPPMI Sensitivity Analysis

This notebook conducts an analysis to evaluate the stability and effectiveness of the tppmi model, focusing on the impact of varying the number of context words. Context words, selected randomly from the most frequently occurring words in the entire corpus, are utilized as the dimensions of the embeddings. These context words remain consistent across all time steps. The analysis explores four different quantities of context words: 200, 500, 1000, and 5000, to determine how the number of context words affects the model's performance.

The timesteps here are one month each

```python
import sys

sys.path.append("packages/TPPMI")
import packages.TPPMI.util as tppmi_util
from embedding_visualization import plot_cosine_similarity_tppmi
```

## Setup & Configurations

```python
y_upper = 0.8 # upper limit for the cosine-plots
```

```python
tppmi_model_200 = tppmi_util.construct_tppmi_from_files(f"data/ppmi-matrices/monthly/200")
tppmi_model_500 = tppmi_util.construct_tppmi_from_files(f"data/ppmi-matrices/monthly/500")
tppmi_model_1000 = tppmi_util.construct_tppmi_from_files(f"data/ppmi-matrices/monthly/1000")
tppmi_model_5000 = tppmi_util.construct_tppmi_from_files(f"data/ppmi-matrices/monthly/5000")
```

## Analysis

### Mass Shootings

#### Overview

- **May:** Uvalde Robb Elementary School shooting (May 24, 2022)
- **November:** Colorado Springs nightclub shooting (November 19, 2022)
- **March:** Nashville school shooting (March 27, 2023)

```python
target_word_shootings = "shooting"
test_words_shootings = ["uvalde", "colorado", "nashville"]
plot_range = [-25, 25]
```

#### 200 context-words

```python
tppmi_model_200.get_tppmi(test_words_shootings + [target_word_shootings]);
```

```python
plot_cosine_similarity_tppmi(target_word_shootings, test_words_shootings, tppmi_model_200, y_upper=y_upper)
```

```python
tppmi_model_500.get_tppmi(test_words_shootings + [target_word_shootings]);
```

```python
plot_cosine_similarity_tppmi(target_word_shootings, test_words_shootings, tppmi_model_500, y_upper=y_upper)
```

#### 1000 context-words

```python
tppmi_model_1000.get_tppmi(test_words_shootings + [target_word_shootings]);
```
